Cat plot,
sns.catplot(x='',y='',data=df,kind=''),
it is came from the factor plot which was deprecated,
9we use catplot insted of factorplot,
kind is used to plot multiple plots in a single plot,

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.catplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',kind='bar')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',kind='strip')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',kind='swarm')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',kind='box')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',kind='violin')
plt.show()
```

```python
sns.catplot(x='size',y='total_bill',data=var,hue='sex',kind='boxen')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=var,hue='sex',kind='point',height=3)
plt.show()
```

```python
sns.catplot(x='day',y='size',data=var,hue='sex',paliette='Accent',kind='count')
plt.show()
```
