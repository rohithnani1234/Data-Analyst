KDE plot
kernel Density Estimation
it creates waves and it describe frequency of data and hoe it flows

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.kdeplot(data=var,x='total_bill',hue='day',multiple='stack')
plt.show()
#multiplt is used for represrnt the data for to color the whole layer, it ca be --stack,layer,fill,,,, default is layer
sns.histplot(data=var,x='total_bill')
plt.show()
sns.kdeplot(data=var,x='total_bill',hue='day',multiple='layer')
plt.show()
sns.kdeplot(data=var,x='total_bill',hue='day',multiple='fill')
plt.show()
```
