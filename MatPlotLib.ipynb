```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("./expense3.xlsx")
df = pd.DataFrame(data)
print(df)
if df['Payment Mode'].isnull().any() or df['Amount'].isnull().any():
    print("Warning: There are missing values in the data.")
    df = df.dropna(subset=['Payment Mode', 'Amount'])
grouped_data = df.groupby('Payment Mode')['Amount'].sum().reset_index()
plt.bar(grouped_data["Payment Mode"], grouped_data["Amount"])
plt.xlabel("Payment Mode")
plt.ylabel("Amount")
plt.title("Total Amount by Payment Mode")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel('./expense3.xlsx')
df = pd.DataFrame(data)

grouped_by = df.groupby("Category")["Amount"].sum()

print(df)
print(grouped_by)

df['Date'] = pd.to_datetime(df['Date'])

fig, ax1 = plt.subplots(2, 1, figsize=(10, 8))

categories = grouped_by.index
amounts = grouped_by.values
ax1[0].bar(range(len(categories)), amounts, tick_label=categories)
ax1[0].set_title('Total Expenses by Category')
ax1[0].set_xlabel('Category')
ax1[0].set_ylabel('Total Amount')

ax1[1].plot(df['Date'], df['Amount'])
ax1[1].set_title('Expenses Over Time')
ax1[1].set_xlabel('Date')
ax1[1].set_ylabel('Amount')

plt.tight_layout()
plt.show()
```
