```python
#pairplot--combining of other plots
#sns.pairplot(df)
#we can easily compare
#by default it will take the all the numberical data and plots the graphs
# we use this plot in machine learning for to represent graph and to compare easily for the model
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.pairplot(var,hue='sex',markers=['*','o'],vars=['total_bill','tip'],hue_order=['Female','Male'],palette='Accent',)
#vars is used to represent the data what we need to need remaining we can neglate
plt.show()
```

```python
sns.pairplot(var,hue='sex',x_vars=['total_bill','tip'],hue_order=['Female','Male'],y_vars=['tip','size'],kind='reg',diag_kind='hist')
'''#x_vars and y_vars are to represent the data what we need to represent acoording to our wish
#kind-- means it like in what graph we can represent they can be reg,kde,hist,scatter, bu default it takes scatter and it behaves as per the data we provide
#diag_kind - is used to represent the grah that which are diagonal or which compares the category that graph can be plooted with severeal kinds like reg,hist,kde,scatter'''
plt.show()
```
