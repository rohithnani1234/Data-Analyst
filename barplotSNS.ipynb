```python
#barplot
#sns.barplot
#is for comparisions
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset('penguins')
var
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=var.island,y=var.bill_length_mm,hue=var.sex,order=order_1,hue_order=["Female","Male"],ci=100,n_boot=3,orient="v",palette='Accent',dodge=True)
#if we given like this then we no need to use data parameter becoz we have already metioned that data of those parameter as var with repect for the data
#order is used for to giving of the arrangemts with our convenint wise , otherwise it show like as per data set metnioned wise
#hue_order is used to keep in order with hue 
#in matplotlib we have to give 2 to three the data for multiple graphing but in seaborn we use hue
# ci-confidence interval-is constructed by taking a percentile interval of the bootstrap distribution. By default errorbar="ci" draws a 95% confidence interval ,is always betwwen in the range of 1 to 100 only
#n_boot is used for number of bootstrap resamples used to estimate the ci and it is optional , the default value attempts to balance time and scalability
#oreint is used to represnt the graph either horizontal or vertical and it will take only if there is numerical data and it does not allow string data or categorical values
#dodge - is used for to represent multiple data as a on data and it will be given in boolean values
plt.show()
```

```python
var=sns.load_dataset('penguins')
sns.barplot(x='island',y='bill_depth_mm',data=var,color='g')
plt.show()
```

```python
sns.set(style="darkgrid") #style must be one of white, dark, whitegrid, darkgrid, ticks
sns.barplot(x='island',y='bill_depth_mm',data=var,palette='icefire',saturation=1,errcolor='red',dodge=False,alpha=0.5)
#saturation is used for to represent dark colors in light colors or smoother colors and its range is between 0 to 1 1 is for true colors
#we used ci to represent error and it can customizes as
#errcolor - to change the color
#errwidth - to change the width of the error
#capsize - to give position the error  ex:capsize=5
#aplha is used to make the color semi-transparent
#the difference between saturation and alpha is saturation will change the color to look smoother but alpha will make that color as semi-transparent
plt.show()
```
