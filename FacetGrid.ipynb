```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
graph=sns.FacetGrid(var,col='day',row='sex',hue='smoker',despine=False,palette='Accent') #
graph.map(plt.scatter,'total_bill','tip',edgecolor='black').add_legend()
plt.show()
#by col and row  we get each category as each plot for every possibility
#add_legend() is used to represent legend of the graphs
```
