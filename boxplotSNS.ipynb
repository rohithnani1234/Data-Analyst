x-axis can be categorical data,
y-axis must be numerical data,
if we need to represent the graph in horizontal then we have to change axis data then it will satisified,

```python
#we categorige the data
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day',y='total_bill',data=var,fill=True,order=['Sun','Thur','Fri','Sat'],showmeans=True)
plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='day',y='total_bill',data=var,fill=True,order=['Sun','Thur','Fri','Sat'],showmeans=True,meanprops={'marker':'+'},linewidth=2,linecolor='brown',)
plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(x='total_bill',y='day',data=var,fill=True,order=['Sun','Thur','Fri','Sat'],showmeans=True,meanprops={'marker':'+'},linewidth=2,linecolor='brown',)
plt.show()
```

```python
sns.set(style='whitegrid')

sns.boxplot(data=var,fill=True,showmeans=True,meanprops={'marker':'+'},linewidth=2,linecolor='brown',orient='h')
plt.show()
#here we did not given any x or y axis , it will represnt  all numerical data on y-axis
```
