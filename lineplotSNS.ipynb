```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,5,1,6,7]
x1=pd.DataFrame({"var":var,"var_1":var_1})
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
sns.lineplot(x="var",y="var_1",data=x1) 
#major diff between seaborn ans matplotlib is here x and y values are will be given string format and they read the values and give labels to it by deault
#we need not to mention the labels exceptiionally
#if we given like x=var and y=var_1 without any string format then it will work like matplotlib without giving any labels
plt.show()
```

```python
#we will load the data withusing seaborn.load_dataset and we dont need to any extension for it
data=sns.load_dataset('penguins').head(30)
data
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=data,hue="sex",size=10,style="sex",palette="Accent",markers=["o",">"],dashes=False,legend=True)
#size=customize the size of the graph
#palette is used to change the color
#marker means it will assign a certain symbol for the point and only for one category
#markers means it will assign a certain symbol for the multiple category and it will given in a list format 
#dashes is for to represent the line in dashes , it will be given in boolean values , auto,brief,full
#we can represent the legend by using legend poarameter, if false so it wont appear 
plt.grid()
plt.title("Penguin Bills",loc='center',)
plt.show()
```
